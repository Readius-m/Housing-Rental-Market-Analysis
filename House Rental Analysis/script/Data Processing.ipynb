## 数据处理

```python
import pandas as pd
import sqlite3
```

读取数据

```python
df = pd.read_csv('北京市租房数据集.csv',encoding='gbk')
df.head(5)
```

初步查看数据

```python
df.shape
```

```python
df.info()
df.describe()
df.duplicated().sum()
df.isnull().sum()
```

数据无空值，重复值

连接sqlite并写入

```python
conn = sqlite3.connect('rental.db')
df.to_sql('rental',conn,index=False,if_exists='replace')
conn.close()
print('数据已成功写入rental.db的rental表')
```
