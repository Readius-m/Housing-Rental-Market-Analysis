## 对每个房源进行评分

```python
import pandas as pd
import numpy as np
from bokeh.core.property import include
```

```python
df = pd.read_csv('北京市租房数据集.csv',encoding='gbk')
df.head(5)
```

复制一份数据进行打分

```python
df_score = df.copy()
df_score.head(5)
```

1.租金评分

```python
df_score['租金评分'] = pd.qcut(df_score['租金'].rank(method='first'),q=5,labels=[5,4,3,2,1]).astype(int)
```

2.面积评分

```python
df_score['面积评分'] = pd.qcut(df_score['面积'].rank(method='first'),q=5,labels=[1,2,3,4,5]).astype(int)
```

3.配套评分（房源周围是否有地铁、学校、医院、商场）

```python
df_score['配套评分'] = (
    df_score['是否有地铁'].apply(lambda x: 1 if x == 1 else 0) +
    df_score['是否有学校'].apply(lambda x: 1 if x == 1 else 0) +
    df_score['是否有医院'].apply(lambda x: 1 if x == 1 else 0) +
    df_score['是否有商场'].apply(lambda x: 1 if x == 1 else 0)
)
df_score['配套分'] = pd.cut(df_score['配套评分'],bins=[0,1,2,3,4,5],labels=[1,2,3,4,5],include_lowest=True).astype(int)
df_score.tail()
```

4.装修评分

```python
df_score['装修情况'].unique()
mapping = {'精装':5,'简装':3,'毛坯':1}
df_score['装修评分'] = df_score['装修情况'].map(mapping).fillna(3).astype(int)
```

5.楼层评分

```python
df_score['所在楼层'].unique()
Floor = {'低楼层':1,'高楼层':1,'中楼层':5}
df_score['楼层评分'] = df_score['所在楼层'].map(Floor).fillna(1).astype(int)
```

性价比总分

```python
df_score['性价比总分'] = (
    df_score['租金评分'] * 0.3 +
    df_score['面积评分'] * 0.2 +
    df_score['配套评分'] * 0.2 +
    df_score['装修评分'] * 0.15 +
    df_score['楼层评分'] * 0.15
)
df_score['性价比总分'].unique()
```

## 每个区域性价比总分的前三位

```python
top_value = df_score.sort_values('性价比总分',ascending=False).groupby('区域').head(3)
top_value[['标题','区域','租金','面积','租金评分','面积评分','性价比总分']]
```

从性价比总分排序结果来看，性价比最高的房源主要集中于通州、大兴、顺义等外围区域，且普遍以“低租金、大面积”的整租或合租户型为主（综合得分达4.5分）；而东城、西城、海淀、朝阳等核心城区虽然地段优越，但因租金评分被大幅拉低，整体性价比得分普遍在3.9至4.1分之间，处于榜单中下游，反映出北京租房市场存在显著的“核心区租金溢价”现象。
